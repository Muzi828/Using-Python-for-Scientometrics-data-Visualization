networkMultiLevel()与networkMultiMode()的区别：
    
    * networkMultiMode()是指标签与另外标签之间的关系，比如author-keyword
    * networkMultiLevel()是指标签与其它标签及自身的关系，比如author-author, keyword-keyword, author-keyword

```python
multi_level = RC1821.networkMultiLevel('C1','SO',stemmer=lambda x:x.split(',')[0])
print(mk.graphStats(multi_level))
```

```python
for item in multi_level.edges.data():
    if 'Univ' in item[0] and 'Univ' in item[1]:
        print(item)
```

```python
multi_mode = RC1821.networkMultiMode('C1','SO',stemmer=lambda x:x.split(',')[0])
print(mk.graphStats(multi_mode))
```

```python
for item in multi_mode.edges.data():
    if 'Univ' in item[0] and 'Univ' in item[1]:
        print(item)
print('没有大学之间的关联关系')
```
