## 探究作者和关键词

```python
two_mode = RC1821.networkTwoMode('DE', 'AF')
print(mk.graphStats(two_mode))
```

数据量上去后，必然会处在一个巨大网络

```python
print(sorted([len(k) for k in nx.connected_components(two_mode)],reverse=True))
```

先进行连线的修剪

```python
mk.dropEdges(two_mode,minWeight=3)
print(mk.graphStats(two_mode))
```

再剔除孤立点

```python
def filter_isolate_nodes(G, min_count=1):
    G_filter = G.subgraph([j for c in nx.connected_components(G) if len(c) > min_count for j in c])
    return G_filter
```

```python
two_mode_filter = filter_isolate_nodes(two_mode,min_count=4)
print(two_mode_filter)
nx.draw_spring(two_mode_filter)
```

```python
dict(two_mode_filter.nodes.data())
```

```python
two_mode_filter.edges.data()
```

```python
node_size = [v['count']*20 for k,v in two_mode_filter.nodes.data()]
edge_width = [v[-1]['weight']*5 for v in two_mode_filter.edges.data()]


plt.figure(figsize=(24,24))

pos = nx.spring_layout(two_mode_filter,seed=98,k=0.4)

nx.draw_networkx_edges(two_mode_filter,pos,width=edge_width,edge_color="#D4D5CE")
nx.draw_networkx_nodes(two_mode_filter,pos,node_size=node_size,edgecolors='k')

G_sub =two_mode_filter.subgraph([k for k,v in two_mode_filter.nodes.data() if v['count']>=50])
label_options = {"ec": "k", "fc": "white", "alpha": 0.7}
nx.draw_networkx_labels(G_sub,pos,font_size=20,bbox=label_options,)


G_sub_2 = two_mode_filter.subgraph([k for k,v in two_mode_filter.nodes.data() if v['count']<50])
nx.draw_networkx_labels(G_sub_2,pos,font_size=10)


plt.savefig('figures/two_mode_DE_and_AF.png',dpi=300,bbox='tight')
```
