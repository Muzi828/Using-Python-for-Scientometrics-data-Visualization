RC.networkCoAuthor()中的参数解释：

    * detailedInfo: [bool]或者可迭代的[WOS标签]数据类型。默认为'False',如果为'True'，所有节点将得到由来自Record对象
        本身的信息组成的信息字符串。这相当于传递列表:['PY'， 'TI'， 'SO'， 'VL'， 'BP']，也可以指定部分WOS标签。
    注: detailedInfo与RC.networkCoCitation()或RC.networkCitation()中的detailedCore使用后返回的结果有所不同，但是实现的功能是类似；
    * weighted: [bool]数据类型。默认为'True'，是否对连线进行加权。如果为'True'，则根据共同作者的数量对连线进行加权；
    * dropNonJournals: [bool]数据类型。默认为'False'，是否从非期刊中删除作者；
    * count: [bool]数据类型。默认为'True'，计算节点的出现次数；
    * useShortNames: [bool]数据类型。默认为'False',如果设置为'True'，则使用作者名称缩写；
    * citeProfile: [bool]数据类型。默认为'False',如果设置为'True'，则会丰富节点属性，添加 'citeProfileCites', 'citeProfileCounts'两个属性，分别记录引文的信息和被引数量。
